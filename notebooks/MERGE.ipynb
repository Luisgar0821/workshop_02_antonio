```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
print(os.getcwd())

# Cargar el dataset
df = pd.read_csv('/mnt/d/Universidad/ETL/workshop_02_antonio/Data/final/merged_music_data.csv')
```

```python
# Configuraci√≥n de gr√°ficos
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

# Vista general
print("üîç Primeras filas:")
display(df.head())

print("\nüìä Informaci√≥n general:")
print(df.info())

print("\nüìà Estad√≠sticas descriptivas:")
display(df.describe(include='all'))

# Valores nulos
print("\nüö® Valores nulos por columna:")
print(df.isnull().sum())

# Distribuci√≥n de popularidad
if 'popularity' in df.columns:
    sns.histplot(df['popularity'].dropna(), bins=20, kde=True)
    plt.title('Distribuci√≥n de popularidad')
    plt.xlabel('Popularidad')
    plt.ylabel('Frecuencia')
    plt.show()

# Duraci√≥n de canciones
if 'duration_ms' in df.columns:
    df['duration_min'] = df['duration_ms'] / 60000
    sns.histplot(df['duration_min'], bins=30, kde=True, color="green")
    plt.title('Distribuci√≥n de duraci√≥n de canciones (minutos)')
    plt.xlabel('Duraci√≥n (min)')
    plt.ylabel('Frecuencia')
    plt.show()

# Canciones m√°s populares
if 'track_name' in df.columns and 'popularity' in df.columns:
    top_tracks = df[['track_name', 'artist_name', 'popularity']].dropna().sort_values(by='popularity', ascending=False).head(10)
    sns.barplot(data=top_tracks, x='popularity', y='track_name', hue='artist_name', dodge=False)
    plt.title('üéß Top 10 canciones m√°s populares')
    plt.xlabel('Popularidad')
    plt.ylabel('Canci√≥n')
    plt.legend(loc='lower right')
    plt.show()

# Canciones con Grammy
if 'won' in df.columns:
    df['won'] = df['won'].fillna('No')
    sns.countplot(x='won', data=df, palette='pastel')
    plt.title('üèÜ Canciones que ganaron Grammy')
    plt.xlabel('Gan√≥ Grammy')
    plt.ylabel('Cantidad')
    plt.show()
```

## Manejo de los valores nulos

Para el manejo de los valores nulos, se seleccionar√≥n las variables que servian para el analisis e interpretaci√≥n, por tanto, la cantidad que se ve no representa una amenaza para el analis e interpretaci√≥n.
